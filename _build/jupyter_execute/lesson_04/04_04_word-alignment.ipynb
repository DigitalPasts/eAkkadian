# L3.9: Exercises

## A. Pronominal Suffixe & Ventive - I
Connect the the morphological elements and translate:

1. *aṭrud* + *šu*
   ```{admonition} Solution
   :class: dropdown  
   *aṭrussu* "I have sent him"
   ```
2. *išpur* + *am* + *kum*
   ```{admonition} Solution
   :class: dropdown  
   *išpurakkum* "He has written you"
   ```
3. *iṭrud* + *am* + *ni*
    ```{admonition} Solution
   :class: dropdown  
   *iṭrudanni* "He has sent me"
   ```
4. *aṭrud* + *kum* + *ši*
   ```{admonition} Solution
   :class: dropdown  
   *aṭrudkušši* "I have sent her to you"
   ```
5. *tašapparā* + *nim* + *ni*
   ```{admonition} Solution
   :class: dropdown  
   *tašapparāninni* "You sent me"
   ```

## B. Pronominal Suffixe & Ventive - II
Translate:
```{admonition} Example
:class: tip
- *išpuranni* = "He/She sent me"
```
1. *tašaparam*
    ```{admonition} Solution
   :class: dropdown  
   *tašapar + am* = "You send to me"
   ```
2. *tašapparānim*
    ```{admonition} Solution
   :class: dropdown  
   *tašapparā + nim* = "You send to me"
   ```
3. *ašapparakkum*
   ```{admonition} Solution
   :class: dropdown  
   *ašappar + am + kum* = "I send to you"
   ```
4. *išapparakkum*
   ```{admonition} Solution
   :class: dropdown  
   *išappar + am + kum* "He sends to you"
   ```
5. *tašapparāšu*
   ```{admonition} Solution
   :class: dropdown  
   *tašapparā + šu* "You send him"
   ```
6. *išapparka*
   ```{admonition} Solution
   :class: dropdown  
   *išappar + ka* "He sends you"
   ```
7. *ašapparkunūšim*
   ```{admonition} Solution
   :class: dropdown  
   *ašappar + kunūšim* "I send you"
   ```
8. *ašpurakkum*
   ```{admonition} Solution
   :class: dropdown  
   *ašpur + am + kum* "I sent to you"
   ```
9. *išpuršunūšim*
   ```{admonition} Solution
   :class: dropdown  
   *išpur + šunūšim* "He/She sent to them"
   ```
10. *ašpurakkunūti*
    ```{admonition} Solution
    :class: dropdown  
    *ašpur + am + kunūti* "I sent you"
    ```

## C. Pronominal Suffixe & Ventive - III
Normalize and translate
```{admonition} Example
:class: tip
- *i-ša-ap-pa-ru-nim* = *išapparūnim* "They send me"
```
1. *iš-pur-šu-nu-ti*
   ```{admonition} Solution
   :class: dropdown  
   *išpuršunūti* "He/She sent them"
   ```
2. *iš-pu-ur-šu*
   ```{admonition} Solution
   :class: dropdown  
   *išpuršu* "He/She sent him*
   ```
3. *am-mi-ni(m) la iš-pu-ra-am*
   ``` {admonition} Tip
   :class: tip
   *ammīnim* = why
   ```
   ```{admonition} Solution
   :class: dropdown  
   *ammīnim lā išpuram* "Why did he not send to me"
   ```
4. *ki-a-am iš-pu-ru-nim*
   ``` {admonition} Tip
   :class: tip
   *kīam* = so, thus
   ```
   ```{admonition} Solution
   :class: dropdown  
   *kīam išpurūnim* "Thus they sent to me"
   ```
5. *šar-ru-um iš-pu-ra-aš-šum*
   ```{admonition} Solution
   :class: dropdown  
   *šarrum išpuraššum* "The king sent him"
   ```
6. *aš-pu-ram*
   ```{admonition} Solution
   :class: dropdown  
   *ašpuram* "I sent there"
   ```
7. *iš-ṭur-šum*
   ```{admonition} Solution
   :class: dropdown  
   *išṭuršum* "He/She wrote him"
   ```
8. *ta-aš₂-ṭu-ra-am*
   ```{admonition} Solution
   :class: dropdown  
   *tašṭuram "You wrote me"
   ```
9. *aš₂-tu₂-ur-šum*
   ```{admonition} Solution
   :class: dropdown  
   *ašṭuršum* "I wrote him"
   ```
10. *iš-ṭu₂-ru-ši-im*
    ```{admonition} Solution
    :class: dropdown  
    *išṭurūšim* "They wrote her"
    ```

## D. Pronominal Suffixe & Ventive - IV
Transliterate, Normalize and translate
1. <font size="11">𒅗𒊭𒁕𒄠</font>
   ```{admonition} Solution
   :class: dropdown  
   *ka-ša-da-am* = *kašādam* = "to arrive"
   ```
2. <font size="11">𒄿𒅗𒀸𒊭𒁕𒄠</font>
   ```{admonition} Solution
   :class: dropdown  
   *i-ka-aš-ša-da-am* = *ikaššadam* = "he/she arrive"
   ```
3. <font size="11">𒄿𒅗𒀸𒊭𒁕𒉏</font>
   ```{admonition} Solution
   :class: dropdown  
   *i-ka-aš-ša-da-nim* = *ikašāddānim* = "they arrive"
   ```
4. <font size="11">𒊭𒁕𒊏𒄠</font>
   ```{admonition} Solution
   :class: dropdown  
   *ša-ṭa-ra-am* = *šaṭaram* = "to write"
   ```
5. <font size="11">𒀀𒊭𒀝𒅗𒀭</font>
   ```{admonition} Solution
   :class: dropdown  
   *a-ša-ak-ka-an* = *ašakkan* = "I place"
   ```
6. <font size="11">𒊭𒅗𒉏</font>
   ```{admonition} Solution
   :class: dropdown  
   *ša-ka-nim* = *šakānim* = "to place"
   ```
7. <font size="11">𒀉𒇻𒌝</font>
   ```{admonition} Solution
   :class: dropdown  
   *eṭ-lu-um* = *eṭlum* = "young man"
   ```
8. <font size="11">𒅗𒀝𒆠𒄿</font>
   ```{admonition} Solution
   :class: dropdown  
   *ka-ak-ki-i* = *kakkī* = "weapons"
   ```
9. <font size="11">𒋛𒎌</font>
   ```{admonition} Solution
   :class: dropdown  
   *SI<sup>meš</sup>* = *qarnātum* = "horns"
   ```
