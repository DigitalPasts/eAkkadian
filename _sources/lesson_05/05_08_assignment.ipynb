## B. Translate the following excerpt from the bilingual royal inscription of Samsu-Iluna (1685-1648 BCE). 
The first section is about Enlil's command to Šamaš to build several temples:

1. *i₃-nu* ᵈEN.LIL₂
2. LUGAL *ša i₃-li₂*
3. *be-lum ra-bi-um*
4. *ša ma-ta₂-tim*
5. *a-na* ᵈUTU *in bu-ni-šu*
6. *dam-qu₂-tim*
7. *ip-pa-al-su₂-ma*
8. ZIMBIR<sup>ki</sup>
9. URU *ṣi-a-tim ma-ha-su₂*
10. BAD₃-*šu e-pe₂-ša-am*
11. E₂.BABBAR *a-na aš-ri-šu*
12. *tu-ur-ra-am* 
...
24. *iq-bi-u₃*

```{admonition} Tip
:class: tip
- *in* = *ina*
- *ippalsu* = (< **palāsum*) "to look, gaze at"
- ZIMBIR<sup>ki</sup> = Sippar (a city)
- BAD₃ = *dūru* "wall"
- E₂.BABBAR = temple name
```

Several lines later Samsu-Iluna tells about his military deeds:

39.  *i₃-nu-šu na-ap-ḫa-ar*
40.  *ma-at šu-me-rim*
41.  *u₃ ak-ka-di-im*
42.  *ša i-ze₂-ru-ni-in-ni*
43.  *in li-ib-bu*
44.  *ša-at-tim iš-ti-a-at
45.  *a-di 8-šu*
46.  *in* <sup>giš</sup>TUKUL *lu a-du-uk*

```{admonition} Tip
:class: tip
- *inūšu* = *at that time, then*
- *iš-ti-a-at* = "one""
- *lū* "verily" 
```
